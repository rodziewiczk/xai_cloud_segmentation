```python
import torch
import torch.nn as nn

class Dw7x7Adapter(nn.Module):
    def __init__(self, in_ch: int = 8, out_ch: int = 5, hidden: int = 128):
        super().__init__()
        self.pw1 = nn.Conv2d(in_ch, hidden, kernel_size=1, bias=True)
        self.dw  = nn.Conv2d(hidden, hidden, kernel_size=7, padding=3, groups=hidden, bias=True)
        self.pw2 = nn.Conv2d(hidden, out_ch, kernel_size=1, bias=True)
        self.act = nn.GELU()

    def forward(self, x):
        x = self.act(self.pw1(x))
        x = self.act(self.dw(x))
        x = self.pw2(x)
        return x

class CloudSENWithAdapter(nn.Module):
    def __init__(self, unet13: nn.Module, adapter: nn.Module,
                 keep_idxs=(1,2,3,4,7,10,11),
                 recon_idxs=(0,5,6,8,9,12)):
        super().__init__()
        self.unet13 = unet13
        self.adapter = adapter
        self.register_buffer("keep_idx_t", torch.tensor(keep_idxs, dtype=torch.long))
        self.register_buffer("tgt_idx_t",  torch.tensor(recon_idxs, dtype=torch.long))

    def forward(self, x8_raw: torch.Tensor) -> torch.Tensor:
        x8_norm = x8_raw
        xk_norm = x8_norm
        xr_norm = self.adapter(xk_norm)
        B, _, H, W = x8_norm.shape
        x13_norm = x8_norm.new_zeros(B, 13, H, W)
        x13_norm.index_copy_(1, self.keep_idx_t, xk_norm)
        x13_norm.index_copy_(1, self.tgt_idx_t, xr_norm)
        return self.unet13(x13_norm)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

base_full_v1 = cloudsen12.load_model_by_name(
    name="UNetMobV2_V1", weights_folder="cloudsen12_models"
).to(device).eval()


base_full_v2 = cloudsen12.load_model_by_name(
    name="UNetMobV2_V2", weights_folder="cloudsen12_models"
).to(device).eval()

normalize_divisor = 10000.0
unet13_core_v1 = getattr(base_full_v1, "model", base_full_v1)
unet13_core_v2 = getattr(base_full_v2, "model", base_full_v2)

adapters_dir = '/content/drive/MyDrive/reconstruction_experiment/'


recon_idxs_list = [
    (1,2,3,7),
    (9,),
    (5,9),
    (0,5,9),
    (0,5,9,10),
    (0,5,9,12),
    (0,5,9,11,12),
    (0,5,9,11,12),
    (0,5,6,9,12),
    (0,4,5,6,9,12),
    (0,4,5,6,8,9,12)]


band_indices_model_dict = {}

def build_model(model, adapter, keep_idxs, recon_idxs):
    return CloudSENWithAdapter(
        unet13=model,
        adapter=adapter,
        keep_idxs=keep_idxs,
        recon_idxs=recon_idxs,
    ).to(device).eval()


def get_adapter_path(band_idxs):
    band_idxs_str = ''.join(map(str,band_idxs))
    fname = f"adapter_{band_idxs_str}_model.pt"
    path = os.path.join(adapters_dir, band_idxs_str, fname)
    return path

for recon_idxs in recon_idxs_list:
    num_bands = len(recon_idxs)
    adapter = Dw7x7Adapter(in_ch=13 - num_bands,
                           out_ch= num_bands,
                           hidden=16
                           ).to(device).eval()
    ckpt = torch.load(get_adapter_path(recon_idxs), map_location=device)
    adapter.load_state_dict(ckpt["state_dict"], strict=False)

    keep_idxs = tuple([i for i in range(13) if i not in recon_idxs])

    model_v1 =  build_model(unet13_core_v1, adapter, keep_idxs, recon_idxs)
    model_v2 =  build_model(unet13_core_v2, adapter, keep_idxs, recon_idxs)

    band_indices_model_dict[recon_idxs] = (model_v1, model_v2)

```

```python
from __future__ import annotations

from typing import Iterable, Tuple, Optional, List, Dict, Any
import numpy as np
import pandas as pd
import torch

def map_to_superclass(arr: np.ndarray, experiment: str) -> np.ndarray:
    if experiment == "cloud_vs_noncloud":
        return ((arr == 1) | (arr == 2)).astype(np.int32)
    if experiment == "cloud_shadow":
        return (arr == 3).astype(np.int32)
    if experiment == "valid_invalid":
        return (arr != 0).astype(np.int32)
    raise ValueError(f"Unknown experiment: {experiment}")


def compute_ip_metrics(gt_mask, pred_mask):
    gt = (gt_mask == 1)
    pred = (pred_mask == 1)

    tp = int(np.logical_and(pred, gt).sum())
    tn = int(np.logical_and(~pred, ~gt).sum())
    fp = int(np.logical_and(pred, ~gt).sum())
    fn = int(np.logical_and(~pred, gt).sum())

    pa = np.nan
    ua = np.nan
    boa = np.nan

    if (tp + fn) > 0:
        pa = tp / (tp + fn)
    if (tp + fp) > 0:
        ua = tp / (tp + fp)
    if (tn + fp) > 0:
        tnr = tn / (tn + fp)
        boa = 0.5 * (pa + tnr)

    if gt.sum() == 0:
        pa = np.nan
        ua = np.nan
        boa = np.nan

    return {"TP": tp, "TN": tn, "FP": fp, "FN": fn, "PA": pa, "UA": ua, "BOA": boa}


def _get_or_build_cache(
    mlstac_path: str,
    normalize_divisor: float,
    device: torch.device,
    load_batch_size: int,
    cache_holder: Optional[List[Dict[str, Any]]],
    cache_dtype: torch.dtype = torch.float32,
) -> Dict[str, Any]:
    if cache_holder and len(cache_holder) == 1:
        c = cache_holder[0]
        if (
            c.get("mlstac_path") == mlstac_path
            and float(c.get("normalize_divisor")) == float(normalize_divisor)
            and c.get("cache_dtype") == str(cache_dtype)
            and c.get("device") == str(device)
        ):
            return c

    meta = mlstac.core.load_metadata(mlstac_path)
    n_scenes = len(meta)

    x_chunks: List[torch.Tensor] = []
    gt_chunks: List[np.ndarray] = []
    scene_ids: List[int] = []

    for start in range(0, n_scenes, load_batch_size):
        end = min(start + load_batch_size, n_scenes)
        batch_meta = meta.iloc[start:end]

        data = mlstac.core.load_data(batch_meta)
        B, C, H, W = data.shape

        x_np = (data[:, :13, :, :] / float(normalize_divisor)).astype(np.float32)
        gt_np = data[:, 13, :, :].astype(np.int32)

        xt = torch.from_numpy(x_np).to(device=device, dtype=cache_dtype, non_blocking=True)
        x_chunks.append(xt)
        gt_chunks.append(gt_np)

        for b in range(B):
            scene_ids.append(int(batch_meta.iloc[b].get("scene_id", start + b)))

    x_all = torch.cat(x_chunks, dim=0)
    gt_all = np.concatenate(gt_chunks, axis=0)

    cache = {
        "mlstac_path": mlstac_path,
        "normalize_divisor": float(normalize_divisor),
        "device": str(device),
        "cache_dtype": str(cache_dtype),
        "x_all_gpu": x_all,
        "gt_all_cpu": gt_all,
        "scene_ids": scene_ids,
        "n_scenes": n_scenes,
    }

    if cache_holder is not None:
        cache_holder[:] = [cache]

    return cache


def run_cloudsen_experiment_mlstac_multi_df(
    model,
    mlstac_path: str,
    bands: Iterable[int],
    experiments: Tuple[str, ...] = ("cloud_vs_noncloud", "cloud_shadow", "valid_invalid"),
    normalize_divisor: float = 10000.0,
    pred_batch_size: int = 10,
    load_batch_size: int = 64,
    cache_holder: Optional[List[Dict[str, Any]]] = None,
    device: Optional[torch.device] = None,
    cache_dtype: torch.dtype = torch.float32,
) -> pd.DataFrame:
    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model = model.to(device).eval()
    bands = list(bands)

    cache = _get_or_build_cache(
        mlstac_path=mlstac_path,
        normalize_divisor=normalize_divisor,
        device=device,
        load_batch_size=load_batch_size,
        cache_holder=cache_holder,
        cache_dtype=cache_dtype,
    )

    x_all_gpu: torch.Tensor = cache["x_all_gpu"]
    gt_all_cpu: np.ndarray = cache["gt_all_cpu"]
    scene_ids: List[int] = cache["scene_ids"]
    n_scenes: int = cache["n_scenes"]

    band_index = torch.tensor(bands, device=device, dtype=torch.long)

    rows: List[Dict[str, Any]] = []

    for start in range(0, n_scenes, pred_batch_size):
        end = min(start + pred_batch_size, n_scenes)

        xt = x_all_gpu[start:end].index_select(dim=1, index=band_index)

        with torch.no_grad():
            logits = model(xt)
            preds = logits.argmax(dim=1)

        preds_batch = preds.detach().cpu().numpy().astype(np.int32)

        for i in range(end - start):
            scene_id = int(scene_ids[start + i])
            pred_mask = preds_batch[i]
            gt_mask = gt_all_cpu[start + i]

            for exp in experiments:
                pred_super = map_to_superclass(pred_mask, exp)
                gt_super = map_to_superclass(gt_mask, exp)
                mets = compute_ip_metrics(gt_super, pred_super)
                rows.append({"scene_id": scene_id, "exp": exp, **mets})

    df = pd.DataFrame(rows)
    desired_cols = ["scene_id", "exp", "TP", "TN", "FP", "FN", "PA", "UA", "BOA"]
    df = df[[c for c in desired_cols if c in df.columns]]
    return df.reset_index(drop=True)


def run_cloudsen_experiment_mlstac_df(
    model,
    cache_holder,
    mlstac_path: str,
    bands: Iterable[int],
    experiment: Optional[str] = None,
    normalize_divisor: float = 10000.0,
    batch_size: int = 32,
) -> pd.DataFrame:
    df = run_cloudsen_experiment_mlstac_multi_df(
        model=model,
        mlstac_path=mlstac_path,
        bands=bands,
        experiments=("cloud_vs_noncloud", "cloud_shadow", "valid_invalid"),
        normalize_divisor=normalize_divisor,
        pred_batch_size=batch_size,
        cache_holder=cache_holder,
        cache_dtype=torch.float32,
    )

    if experiment is None:
        return df

    return df[df["exp"] == experiment].reset_index(drop=True)
```

```python
import numpy as np
import pandas as pd
from scipy import stats

CI_LABEL = "90"
ALPHA = 1.0 - float(CI_LABEL) / 100.0

def compute_row(
    df_full: pd.DataFrame,
    df_adapter: pd.DataFrame,
    band_idxs,
    group_size: int = 5,
) -> pd.DataFrame:
    metrics = ["BOA", "PA", "UA"]

    df_join = df_full.merge(
        df_adapter,
        how="inner",
        on=["scene_id", "exp"],
        suffixes=("_full", "_adapter"),
    ).copy()

    df_join["roi_id"] = (df_join["scene_id"].astype(int) // int(group_size))

    agg_cols = {f"{m}_full": "mean" for m in metrics}
    agg_cols |= {f"{m}_adapter": "mean" for m in metrics}

    df_roi = (
        df_join
        .groupby(["exp", "roi_id"], dropna=False, as_index=False)
        .agg(agg_cols)
    )

    rows_dict = {}
    for exp, dfE in df_roi.groupby("exp", dropna=False):
        for metric in metrics:
            diffs = (dfE[f"{metric}_adapter"] - dfE[f"{metric}_full"]).to_numpy(dtype=float)
            diffs = diffs[np.isfinite(diffs)]
            n = int(diffs.size)

            mean_diff = float(np.mean(diffs)) if n else np.nan

            if n >= 2:
                sd = float(np.std(diffs, ddof=1))
                se = sd / np.sqrt(n)
                t_crit = float(stats.t.ppf(1.0 - ALPHA / 2.0, df=n - 1))
                ci_low = mean_diff - t_crit * se
                ci_high = mean_diff + t_crit * se
            else:
                ci_low = np.nan
                ci_high = np.nan

            baseline_vals = dfE[f"{metric}_full"].to_numpy(dtype=float)
            baseline_vals = baseline_vals[np.isfinite(baseline_vals)]
            baseline_mean = float(np.mean(baseline_vals)) if baseline_vals.size else np.nan

            prefix = f"{exp}_d{metric}"
            rows_dict[f"{prefix}_n_roi"] = n
            rows_dict[f"{prefix}_mean"] = mean_diff
            rows_dict[f"{prefix}_ci{CI_LABEL}_low"] = ci_low
            rows_dict[f"{prefix}_ci{CI_LABEL}_high"] = ci_high
            rows_dict[f"{prefix}_baseline_mean_full"] = baseline_mean

    return pd.DataFrame([rows_dict], index=[str(band_idxs)])

def build_diff_df(experiment_results_dict, bands_full):
    df_full_v1, df_full_v2 = experiment_results_dict[bands_full]

    rows_v1 = []
    rows_v2 = []
    for band_idxs, (df_v1, df_v2) in experiment_results_dict.items():
        rows_v1.append(compute_row(df_full_v1, df_v1, band_idxs))
        rows_v2.append(compute_row(df_full_v2, df_v2, band_idxs))

    return pd.concat(rows_v1, axis=0), pd.concat(rows_v2, axis=0)

def compact_relative_ci_df(df_raw: pd.DataFrame, ci_label: str = CI_LABEL, digits: int = 2) -> pd.DataFrame:
    df = df_raw.copy()

    prefixes = []
    for c in df.columns:
        if c.endswith("_mean") and "_d" in c:
            prefixes.append(c[:-5])

    for prefix in prefixes:
        mean_col = prefix + "_mean"
        low_col = prefix + f"_ci{ci_label}_low"
        high_col = prefix + f"_ci{ci_label}_high"
        base_col = prefix + "_baseline_mean_full"

        if not all(col in df.columns for col in [mean_col, low_col, high_col, base_col]):
            continue

        denom = pd.to_numeric(df[base_col], errors="coerce").replace(0, np.nan)
        mean = pd.to_numeric(df[mean_col], errors="coerce")
        low = pd.to_numeric(df[low_col], errors="coerce")
        high = pd.to_numeric(df[high_col], errors="coerce")

        mean_pct = 100.0 * mean / denom
        delta_pct = 50.0 * (high - low) / denom

        df[prefix + "_mean_pct"] = mean_pct
        df[prefix + f"_delta{ci_label}_pct"] = delta_pct
        df[prefix + f"_pretty{ci_label}"] = (
            mean_pct.round(digits).astype(str) + "±" + delta_pct.round(digits).astype(str)
        )

    keep = []
    for prefix in prefixes:
        ncol = prefix + "_n_roi"
        pcol = prefix + f"_pretty{ci_label}"
        if pcol in df.columns:
            keep.append(pcol)
        if ncol in df.columns:
            keep.append(ncol)

    return df[keep]

mlstac_path_val_v1 = "/content/drive/MyDrive/data/cloudsen12_data/cloudsen12_val/val_509_high.mlstac"
mlstac_path_val_v2 = "/content/drive/MyDrive/data/cloudsen12plus_data/cloudsen12plus_val/validation_509_high.mlstac"
mlstac_paths = (mlstac_path_val_v1, mlstac_path_val_v2)

SAVED_VAL = [[], []]
experiment_results_dict = {}

def run_for_models(models, mlstac_paths, bands, cache_holders):
    return tuple(
        run_cloudsen_experiment_mlstac_df(
            model=model,
            cache_holder=cache_holder,
            mlstac_path=path,
            bands=bands,
        )
        for model, path, cache_holder in zip(models, mlstac_paths, cache_holders)
    )

for band_idxs, models in band_indices_model_dict.items():
    bands = [i for i in range(13) if i not in band_idxs]
    experiment_results_dict[band_idxs] = run_for_models(models, mlstac_paths, bands, SAVED_VAL)

bands_full = tuple(range(13))
models_full = (unet13_core_v1, unet13_core_v2)
experiment_results_dict[bands_full] = run_for_models(models_full, mlstac_paths, bands_full, SAVED_VAL)

df_v1_raw, df_v2_raw = build_diff_df(experiment_results_dict, bands_full)
df_v1_pretty_rel = compact_relative_ci_df(df_v1_raw)
df_v2_pretty_rel = compact_relative_ci_df(df_v2_raw)
```

```python
import numpy as np
import torch
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
import mlstac

SEED = 123
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

mpl.rcParams.update({
    "figure.dpi": 150,
    "savefig.dpi": 300,
    "font.family": "serif",
    "font.serif": ["CMU Serif", "Computer Modern Roman", "Latin Modern Roman", "DejaVu Serif"],
    "mathtext.fontset": "cm",
    "axes.unicode_minus": False,
    "font.size": 12,
    "axes.titlesize": 14,
    "axes.labelsize": 13,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11,
    "legend.fontsize": 10,
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
})

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
normalize_divisor = 10000.0

PERCENTILES = (50, 75, 90, 95, 99)
BANDS_FULL = tuple(range(13))

RECON_LIST = [
    (0, 4, 5, 6, 9, 12),
    (1, 2, 3, 7),
]

cmap = ListedColormap(["#0000FF", "#FFFFFF", "#BFBFBF", "#000000"])
legend_handles = [
    mpatches.Patch(color=cmap.colors[0], label="klasa 0 – czyste niebo"),
    mpatches.Patch(color=cmap.colors[1], label="klasa 1 – gęsta chmura"),
    mpatches.Patch(color=cmap.colors[2], label="klasa 2 – cienka chmura"),
    mpatches.Patch(color=cmap.colors[3], label="klasa 3 – cień chmury"),
]

def _stretch01(ch: np.ndarray, lo=2.0, hi=98.0, eps=1e-8) -> np.ndarray:
    a = np.percentile(ch, lo)
    b = np.percentile(ch, hi)
    return np.clip((ch - a) / (b - a + eps), 0.0, 1.0)

def make_rgb(x13: np.ndarray) -> np.ndarray:
    r = _stretch01(x13[3])
    g = _stretch01(x13[2])
    b = _stretch01(x13[1])
    return np.stack([r, g, b], axis=-1)

def strip_axes_keep_frame(ax):
    ax.set_xticks([])
    ax.set_yticks([])
    ax.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)

def blue_frame(ax, lw=2.0):
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_edgecolor("#0000FF")
        spine.set_linewidth(lw)

def maybe_cuda_cleanup():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

@torch.no_grad()
def disagreement_rate_from_logits(logits_p: torch.Tensor, logits_q: torch.Tensor) -> torch.Tensor:
    yp = torch.argmax(logits_p, dim=1)
    yq = torch.argmax(logits_q, dim=1)
    return (yp != yq).float().mean(dim=(1, 2))

def pick_percentile_scenes(values: np.ndarray, meta, percentiles=PERCENTILES):
    picks = []
    for p in percentiles:
        q = np.percentile(values, p)
        idx = int(np.argmin(np.abs(values - q)))
        scene_id = int(meta.iloc[idx].get("scene_id", idx))
        picks.append((p, idx, scene_id, float(values[idx])))
    return picks

def recon_tag(recon_idxs):
    return "".join(map(str, recon_idxs))

def title_adapter(n_keep: int):
    return f"(a) Predykcja: model z adapterem ({n_keep} pasm)"

def run_variant(*, name: str, model_full, model_adpt, mlstac_path: str,
                out_prefix: str, recon_idxs: tuple, batch_size: int = 1):

    keep_idxs = tuple([i for i in range(13) if i not in recon_idxs])
    n_keep = len(keep_idxs)

    model_full = model_full.to(device).eval()
    model_adpt = model_adpt.to(device).eval()

    meta = mlstac.core.load_metadata(mlstac_path)
    n_scenes = len(meta)
    vals = np.zeros(n_scenes, dtype=np.float64)

    for start in range(0, n_scenes, batch_size):
        end = min(start + batch_size, n_scenes)
        batch_meta = meta.iloc[start:end]
        data = mlstac.core.load_data(batch_meta)

        x13 = (data[:, BANDS_FULL, :, :] / float(normalize_divisor)).astype(np.float32)
        xt13 = torch.from_numpy(x13).to(device, non_blocking=True)

        with torch.inference_mode():
            logits_full = model_full(xt13)
            logits_adpt = model_adpt(xt13[:, keep_idxs])

        vals[start:end] = disagreement_rate_from_logits(logits_full, logits_adpt).detach().cpu().numpy()

        del data, x13, xt13, logits_full, logits_adpt
        maybe_cuda_cleanup()

    picks = pick_percentile_scenes(vals, meta, PERCENTILES)
    print(f"\n[{name} | recon={recon_idxs} | keep={n_keep}] wybrane sceny (p, idx, scene_id, disagreement):")
    for p, idx, scene_id, v in picks:
        print(f"p{p}: idx={idx}, scene_id={scene_id}, disagreement={v:.6f}")

    rtag = recon_tag(recon_idxs)
    for p, idx, scene_id, metric_value in picks:
        one = mlstac.core.load_data(meta.iloc[[idx]])

        x13 = (one[0, BANDS_FULL] / float(normalize_divisor)).astype(np.float32)
        gt = one[0, 13].astype(np.int32)
        rgb = make_rgb(x13)

        xt13 = torch.from_numpy(x13[None]).to(device, non_blocking=True)

        with torch.inference_mode():
            pred_full = torch.argmax(model_full(xt13), dim=1).squeeze(0).cpu().numpy().astype(np.int32)
            pred_adpt = torch.argmax(model_adpt(xt13[:, keep_idxs]), dim=1).squeeze(0).cpu().numpy().astype(np.int32)

        fig, axes = plt.subplots(
            2, 2,
            figsize=(9.0, 6.9),
            gridspec_kw={"wspace": 0.03, "hspace": 0.14}
        )
        (ax_tl, ax_tr), (ax_bl, ax_br) = axes

        ax_tl.imshow(pred_adpt, cmap=cmap, vmin=0, vmax=3, interpolation="nearest")
        ax_tl.set_title(title_adapter(n_keep), pad=10)
        strip_axes_keep_frame(ax_tl)

        ax_tr.imshow(pred_full, cmap=cmap, vmin=0, vmax=3, interpolation="nearest")
        ax_tr.set_title("(b) Predykcja: model pełny (13 pasm)", pad=10)
        strip_axes_keep_frame(ax_tr)

        ax_bl.imshow(rgb)
        ax_bl.set_title("(c) Kompozycja RGB (Sentinel-2)", pad=10)
        ax_bl.axis("off")

        ax_br.imshow(gt, cmap=cmap, vmin=0, vmax=3, interpolation="nearest")
        ax_br.set_title("(d) Maska referencyjna (GT)", pad=10)
        strip_axes_keep_frame(ax_br)

        blue_frame(ax_tl, lw=2.0)
        blue_frame(ax_tr, lw=2.0)
        blue_frame(ax_br, lw=2.0)

        ax_br.legend(
            handles=legend_handles,
            loc="lower right",
            frameon=True,
            framealpha=0.95,
            borderpad=0.3,
            handlelength=1.1,
            handletextpad=0.4,
            fontsize=10,
        )

        fig.subplots_adjust(left=0.03, right=0.99, top=0.95, bottom=0.06)

        out_pdf = f"{out_prefix}_recon{rtag}_p{p}_idx{idx:04d}.pdf"
        fig.savefig(out_pdf, bbox_inches="tight", pad_inches=0.03)
        plt.show()
        plt.close(fig)

        del one, x13, gt, rgb, xt13, pred_full, pred_adpt, fig, axes
        maybe_cuda_cleanup()

    model_full.cpu()
    model_adpt.cpu()
    maybe_cuda_cleanup()

mlstac_path_final = (
    "/content/drive/MyDrive/data/cloudsen12_data/cloudsen12_test/test_509_high.mlstac",
    "/content/drive/MyDrive/data/cloudsen12plus_data/cloudsen12plus_test_high.mlstac",
)

for recon_idxs in RECON_LIST:
    if recon_idxs not in band_indices_model_dict:
        raise KeyError(f"Brak modelu adaptera w band_indices_model_dict dla recon_idxs={recon_idxs}")

    model_adpt_v1, model_adpt_v2 = band_indices_model_dict[recon_idxs]

    variants = [
        dict(
            name="v1",
            model_full=unet13_core_v1,
            model_adpt=model_adpt_v1,
            mlstac_path=mlstac_path_final[0],
            out_prefix="unetmobv2_v1_cloudsen12",
            recon_idxs=recon_idxs,
        ),
        dict(
            name="v2",
            model_full=unet13_core_v2,
            model_adpt=model_adpt_v2,
            mlstac_path=mlstac_path_final[1],
            out_prefix="unetmobv2_v2_cloudsen12plus",
            recon_idxs=recon_idxs,
        ),
    ]

    for v in variants:
        run_variant(**v, batch_size=1)
```
